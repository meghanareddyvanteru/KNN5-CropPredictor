```python
#Adding dependencies
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
#loading data set into pandas
crop_data=pd.read_csv('/content/Crop_recommendation.csv')
crop_data.head()
```

```python
#number of rows and columns
crop_data.shape
```

```python
crop_data.describe()
```

```python
crop_data['label'].value_counts()
```

```python
crop_data.groupby('label').mean()
```

```python
#separating data and labels
x=crop_data.drop(columns='label',axis=1)
y=crop_data['label']
print(x)
print(y)
```

```python
#training and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,stratify=y,random_state=42)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#Model Training-->KNeighborsClassifier
knn_model=KNeighborsClassifier(n_neighbors=5)
#training KNeighborsClassifier with training data
knn_model.fit(x_train,y_train)
```

```python
y_pred=knn_model.predict(x_test)
```

```python
#model evaluation
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy on test data:",accuracy)
```

```python
y_train_predict=knn_model.predict(x_train)
training_data_accuracy=accuracy_score(y_train_predict,y_train)
print("Accuracy on training data:",training_data_accuracy)
```

```python
#making a predictive system
input_data=(52,71,16,27.74274761,68.53997144,7.0758864720000005,71.78615328)
input=np.array(input_data)
feature_names=["N","P","K","temperature","humidity","ph","rainfall"]
input_prediction=pd.DataFrame(data=[input],columns=feature_names)
crop=knn_model.predict(input_prediction)
print(crop)
```
